```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# CSV   DataFrame 
file_path = '../csv/united_summer2015.csv'  #     .
data = pd.read_csv(file_path, delimiter=',')

#  50, 100, 200  
sample_sizes = [50, 100, 200]

for size in sample_sizes:
    #  
    sample_data = data.sample(n=size, random_state=42)  # random_state    .

    #      
    average_delay = sample_data['Delay'].mean()
    variance_delay = sample_data['Delay'].var()

    #     
    plt.figure(figsize=(10, 6))
    plt.hist(sample_data['Delay'], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Flight Delay Distribution (Sample Size: {size})')
    plt.xlabel('Delay Time')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    #  
    print(f"Sample Size: {size}")
    print(f"  : {average_delay}")
    print(f"  : {variance_delay}")
    print("-" * 30)
```

```python
import pandas as pd
import numpy as np

# CSV   DataFrame 
file_path = '../csv/united_summer2015.csv'  #     .
data = pd.read_csv(file_path, delimiter=',')

#  
sample_size = 100
data1 = data.sample(n=sample_size, replace=True,)
#  
sampling_count = 100

# 1000    
sample_means = []

# 1000 
for _ in range(sampling_count):
    #    
    sample_data = data1['Delay'].sample(n=sample_size, replace=True)

    #     
    sample_means.append(sample_data.mean())

# 1000     
overall_mean = np.mean(sample_means)
overall_variance = np.var(sample_means)

#  
print(f"1000    : {overall_mean}")
print(f"1000    : {overall_variance}")
```
