위에서 읽어 들인 데이터에서 50개, 100개, 200개의 샘플을 랜덤하게 추출해서 각각의 샘플에 대해 비행기 출발 지연 시간의 평균과 분산을 구하고 비행지연 시간의  분포를 그래프로 그려주는 파이썬 코드를 만들어줘

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# CSV 파일을 읽어서 DataFrame 생성
file_path = 'united_summer2015.csv'  # 파일 경로를 적절히 수정해야 합니다.
data = pd.read_csv(file_path, delimiter=',')

# 랜덤하게 50개, 100개, 200개의 샘플 추출
sample_sizes = [50, 100, 200]

for size in sample_sizes:
    # 샘플 추출
    sample_data = data.sample(n=size, random_state=42)  # random_state는 재현성을 위한 시드 값입니다.

    # 출발 지연 시간의 평균과 분산 계산
    average_delay = sample_data['Delay'].mean()
    variance_delay = sample_data['Delay'].var()

    # 지연 시간의 분포를 그래프로 그리기
    plt.figure(figsize=(10, 6))
    plt.hist(sample_data['Delay'], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Flight Delay Distribution (Sample Size: {size})')
    plt.xlabel('Delay Time')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    # 결과 출력
    print(f"Sample Size: {size}")
    print(f"평균 지연 시간: {average_delay}")
    print(f"지연 시간 분산: {variance_delay}")
    print("-" * 30)
```

원본 데이터에서 100개의 샘플 데이터를 추출한 뒤, 이 추출된 데이터에서 100개를 다시 복원 추출하는 작업을 1,00번 하면서 평균을 구한다. 이렇게 생성된 100개 평균의 평균과 분산은 어떻게 될까?

```python
# 4-1

import pandas as pd
import numpy as np

# CSV 파일을 읽어서 DataFrame 생성
file_path = 'united_summer2015.csv'  # 파일 경로를 적절히 수정해야 합니다.
data = pd.read_csv(file_path, delimiter=',')

# 샘플 사이즈
sample_size = 100
data1 = data.sample(n=sample_size, replace=True, random_state=42)
# 샘플링 횟수
sampling_count = 100

# 100번 샘플링한 결과를 저장할 리스트
sample_means = []

# 100번 샘플링
for _ in range(sampling_count):
    # 복원 추출로 샘플 추출
    sample_data = data1['Delay'].sample(n=sample_size, replace=True)

    # 샘플의 평균 계산 및 저장
    sample_means.append(sample_data.mean())

# 1000번 샘플링한 결과의 평균과 분산 계산
overall_mean = np.mean(sample_means)
overall_variance = np.var(sample_means)

# 결과 출력
print(f"100번 각각의 샘플 평균의 평균: {overall_mean}")
print(f"100번 각각의 샘플 평균의 분산: {overall_variance}")
```

```python
# 4-2

import pandas as pd
import numpy as np

# CSV 파일을 읽어서 DataFrame 생성
file_path = 'united_summer2015.csv'  # 파일 경로를 적절히 수정해야 합니다.
data = pd.read_csv(file_path, delimiter=',')

# 샘플 사이즈
sample_size = 100
data1 = data.sample(n=sample_size, replace=True, random_state=42)
# 샘플링 횟수
sampling_count = 100

# 100번 샘플링한 결과를 저장할 리스트
sample_means = []

# 100번 샘플링
for _ in range(sampling_count):
    # 복원 추출로 샘플 추출
    sample_data = data['Delay'].sample(n=sample_size, replace=True)

    # 샘플의 평균 계산 및 저장
    sample_means.append(sample_data.mean())

# 1000번 샘플링한 결과의 평균과 분산 계산
overall_mean = np.mean(sample_means)
overall_variance = np.var(sample_means)

# 결과 출력
print(f"100번 각각의 샘플 평균의 평균: {overall_mean}")
print(f"100번 각각의 샘플 평균의 분산: {overall_variance}")
```
