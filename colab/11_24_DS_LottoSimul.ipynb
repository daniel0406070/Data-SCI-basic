## 통계학이란? 각종 분포는 왜 사용하나?

**베르누이 시행**(Bernoulli trial 또는 binomial trial)은 확률론과 통계학에서 임의의 결과가 '성공' 또는 '실패'의 두 가지 중 하나인 실험을 뜻한다. 다시 말해 '예' 또는 '아니오' 중 하나의 결과를 낳는 실험을 말한다.

예를 들면, 하나의 시행은 다음과 같은 질문에 답할 수 있는 실험이다.

* 동전이 앞면이 위를 향하고 있는가?
* 새로 태어난 아기가 여자인가?
* 사람의 눈이 녹색인가?
*모기가 해당영역에 살충제가 뿌려진 후 죽었는가?
* 소비자가 물건을 구입하였는가?
* 시민이 어느 특정 후보에게 투표하였는가?


그렇기 때문에 반드시 '성공'과 '실패'가 아니라 하여도, 가능한 결과가 두 가지라면 되는 것이다.

## 이항 분포(二項分布)는 연속된 n번의 독립적 시행에서 각 시행이 확률 p를 가질 때의 이산 확률 분포이다. 이러한 시행은 베르누이 시행이라고 불리기도 한다. 사실, n=1일 때 이항 분포는 베르누이 분포이다.

기본적인 예: 일반적인 주사위를 10회 던져서 숫자 6이 나오는 횟수를 센다. 이 분포는 n = 10이고 p = 1/6인 이항분포이다.

다른 예로는, 아주 많은 인구의 5%가 쌍꺼풀이 있다고 해보자. 그리고 100명을 무작위적으로 선택한다. 당신이 선택한 쌍꺼풀을 가진 사람의 수는 n = 100이고 p = 0.05인 이항분포를 따른다.

## 로또 번호(1~45)를 무작위로 sampling with replacement로 50번, 100번, 1,000번 샘플링 하는 것을 5,0000번 진행할 때 로또번호 숫자 중 짝수의 숫자가 나오는 횟수를 히스토그램으로 보이시오.

```python
import random
import matplotlib.pyplot as plt

# 반복 횟수 설정
iteration_count = 50000

# 샘플링 횟수 설정
sample_counts = [50, 100, 1000]

# 짝수의 숫자가 나오는 횟수를 저장할 리스트 초기화
even_counts = {50: [], 100: [], 1000: []}

# 각 샘플링 횟수에 대해 시뮬레이션 수행
for sample_count in sample_counts:
    current_even_counts = []

    # 시뮬레이션 수행
    for _ in range(iteration_count):
        # 1부터 45까지의 로또 번호 중 무작위로 선택 (복원 추출)
        lotto_numbers = random.choices(range(1, 46), k=sample_count)

        # 짝수의 숫자가 몇 번 나왔는지 확인 후 리스트에 추가
        even_count = sum(1 for num in lotto_numbers if num % 2 == 0)
        current_even_counts.append(even_count)

    # 결과 리스트에 현재 결과를 추가
    even_counts[sample_count] = current_even_counts

# 히스토그램 그리기
plt.figure(figsize=(15, 5))

for i, sample_count in enumerate(sample_counts):
    plt.subplot(3, 1, i + 1)
    plt.hist(even_counts[sample_count], bins=range(sample_count + 2), align='left', edgecolor='black')
    plt.xlabel(f'Number of Even Numbers in {sample_count} Lotto Samples')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Even Numbers in {iteration_count} Lotto Samples of {sample_count} Numbers Each (With Replacement)')

plt.tight_layout()
plt.show()
```

## 로또 번호에대해 100,000번의 반복 동안 샘플링 횟수를 100번, 500번, 1000번 실시했을때 숫자 23이 나오는 횟수를 기록하고, 나온 횟수를 0~100까지의 범위로 제한하여 히스토그램으로 표시하는 코드를 작성하시오.

```python
import random
import matplotlib.pyplot as plt

# 반복 횟수 설정
iteration_count = 100000

# 샘플링 횟수 설정
sample_counts = [50, 100, 500, 1000]

# 숫자 23이 나오는 횟수를 저장할 리스트 초기화
count_of_23 = []

# 각 샘플링 횟수에 대해 시뮬레이션 수행
for sample_count in sample_counts:
    current_count_of_23 = []

    # 시뮬레이션 수행
    for _ in range(iteration_count):
        # 1부터 45까지의 로또 번호 중 무작위로 선택 (복원 추출)
        lotto_numbers = random.choices(range(1, 46), k=sample_count)

        # 숫자 23이 몇 번 나왔는지 확인 후 리스트에 추가
        current_count_of_23.append(lotto_numbers.count(23))

    # 23이 나오는 횟수가 100 이상인 경우 100으로 설정
    current_count_of_23 = [min(count, 100) for count in current_count_of_23]

    # 결과 리스트에 현재 결과를 추가
    count_of_23.append(current_count_of_23)

# 히스토그램 그리기
plt.figure(figsize=(15, 5))

for i, sample_count in enumerate(sample_counts):
    #plt.subplot(1, 4, i + 1)
    plt.subplot(4, 1, i + 1)
    plt.hist(count_of_23[i], bins=range(101), align='left', edgecolor='black')
    plt.xlabel('Number of Times 23 Appeared (Capped at 100)')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of the Number 23 in {iteration_count} Lotto Samples of {sample_count} Numbers Each (With Replacement)')

plt.tight_layout()
plt.show()
```

## 로또 번호(1~45)를 무작위로 sampling with replacement로 50번, 100번, 1,000번 샘플링 하는 것을 5,0000번 진행할 때 로또번호 숫자 중 짝수의 숫자가 나오는 횟수를 히스토그램으로 보이고 각 히스토그램의 평균과 분산도 구하시오.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# 반복 횟수 설정
iteration_count = 50000

# 샘플링 횟수 설정
sample_counts = [50, 100, 1000]

# 짝수의 숫자가 나오는 횟수를 저장할 리스트 초기화
even_counts = {50: [], 100: [], 1000: []}

# 각 샘플링 횟수에 대해 시뮬레이션 수행
for sample_count in sample_counts:
    current_even_counts = []

    # 시뮬레이션 수행
    for _ in range(iteration_count):
        # 1부터 45까지의 로또 번호 중 무작위로 선택 (복원 추출)
        lotto_numbers = random.choices(range(1, 46), k=sample_count)

        # 짝수의 숫자가 몇 번 나왔는지 확인 후 리스트에 추가
        even_count = sum(1 for num in lotto_numbers if num % 2 == 0)
        current_even_counts.append(even_count)

    # 결과 리스트에 현재 결과를 추가
    even_counts[sample_count] = current_even_counts

# 히스토그램 그리기
plt.figure(figsize=(15, 5))

for i, sample_count in enumerate(sample_counts):
    plt.subplot(3, 1, i + 1)
    plt.hist(even_counts[sample_count], bins=range(sample_count + 2), align='left', edgecolor='black')
    plt.xlabel(f'Number of Even Numbers in {sample_count} Lotto Samples')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Even Numbers in {iteration_count} Lotto Samples of {sample_count} Numbers Each')

    # 각 히스토그램의 평균과 분산 계산
    mean_value = np.mean(even_counts[sample_count])
    variance_value = np.var(even_counts[sample_count])

    print(f"Sample Count: {sample_count}")
    print(f"Mean: {mean_value:.2f}")
    print(f"Variance: {variance_value:.2f}\n")

plt.tight_layout()
plt.show()
```

## 로또 숫자  1~45까지 숫자 중 6개를 sampling without replacement 로 샘플링한 뒤 숫자의 합을 구하고 그 작업을 1000번, 10,000번, 100,000번 반복하는 시뮬레이션을 하고 결과를 그래프로 보여주는 파이썬 코드를 작성하시오.

```python
import random
import matplotlib.pyplot as plt

# 시뮬레이션 설정
simulation_counts = [1000, 10000, 100000]

# 각 시뮬레이션 별 결과를 저장할 리스트 초기화
results = []

# 각 시뮬레이션 수행
for simulation_count in simulation_counts:
    # 현재 시뮬레이션의 결과를 저장할 리스트 초기화
    current_results = []

    # 시뮬레이션 수행
    for _ in range(simulation_count):
        # 1부터 45까지의 숫자 중에서 중복 없이 6개 샘플링
        lotto_numbers = random.sample(range(1, 46), 6)

        # 샘플링된 숫자의 합을 계산하여 리스트에 추가
        current_results.append(sum(lotto_numbers))

    # 현재 시뮬레이션 결과를 전체 결과 리스트에 추가
    results.append(current_results)

# 그래프 그리기
plt.figure(figsize=(12, 8))

for i, simulation_count in enumerate(simulation_counts):
    plt.subplot(3, 1, i + 1)
    plt.hist(results[i], bins=range(17, 255), edgecolor='black')
    plt.xlabel('Sum of Lotto Numbers')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Lotto Number Sums\n({simulation_count} Simulations)')

plt.tight_layout()
plt.show()
```

## 로또 숫자  1~45까지 숫자 중 6개를 sampling without replacement 로 샘플링한 뒤 숫자의 합을 구하고 그 작업을 1000번, 10,000번, 100,000번 반복하고 이때 각각(1,000번, 10,000번, 100,000번)의 평균과 분산 값도 계산하여  결과를 보여주는 그래프를 작성하시오.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# 반복 횟수 설정
iteration_counts = [1000, 10000, 100000]

# 각 반복 횟수에 대해 시뮬레이션 수행
for iteration_count in iteration_counts:
    sums = []

    # 시뮬레이션 수행
    for _ in range(iteration_count):
        # 1부터 45까지의 로또 번호 중 무작위로 6개 선택 (비복원 추출)
        lotto_numbers = random.sample(range(1, 46), k=6)

        # 선택된 6개의 숫자의 합을 계산하여 리스트에 추가
        total_sum = sum(lotto_numbers)
        sums.append(total_sum)

    # 평균과 분산 계산
    mean_value = np.mean(sums)
    variance_value = np.var(sums)

    # 히스토그램 그리기
    plt.hist(sums, bins=range(17, 255), align='left', edgecolor='black')
    plt.xlabel('Sum of 6 Lotto Numbers')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of the Sum of 6 Lotto Numbers ({iteration_count} Iterations)\nMean: {mean_value:.2f}, Variance: {variance_value:.2f}')
    plt.show()
```

## 로또 숫자 1~45까지 숫자 중 6개를 sampling without replacement 로 샘플링한 뒤 숫자의 합을 구하는 작업을 50번, 100번, 200번씩 한다고 하자.  그 샘플링 작업을 10,000번 반복할 때 각각(50번, 100번, 200번 샘플링)의 평균과 분산 값을 계산하여 결과를 그래프로 보여주는 파이썬 코드를 작성하시오.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# 반복 횟수 설정
iteration_count = 10000

# 샘플링 횟수 설정
sample_counts = [50, 100, 200]

# 각 샘플링 횟수에 대해 시뮬레이션 수행
for sample_count in sample_counts:
    sums = []

    # 시뮬레이션 수행
    for _ in range(iteration_count):
        # 1부터 45까지의 로또 번호 중 무작위로 6개 선택 (비복원 추출)
        lotto_numbers = random.sample(range(1, 46), k=6)

        # 선택된 6개의 숫자의 합을 계산하여 리스트에 추가
        total_sum = sum(lotto_numbers)
        sums.append(total_sum)

    # 평균과 분산 계산
    mean_value = np.mean(sums)
    variance_value = np.var(sums)

    # 히스토그램 그리기
    plt.hist(sums, bins=range(7, 252), align='left', edgecolor='black')
    plt.xlabel('Sum of 6 Lotto Numbers')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of the Sum of 6 Lotto Numbers ({iteration_count} Iterations, {sample_count} Samples Each)\nMean: {mean_value:.2f}, Variance: {variance_value:.2f}')
    plt.show()
```

## 로또 번호(1~45)에대해 복원 샘플링을 50번 실시 했을 때 숫자 23이 나오는 횟수 카운팅 할때 이를 10000번 반복했을때  결과를 그래프로 보여주는 파이썬 코드를 작성하시오.
