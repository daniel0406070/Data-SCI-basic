```python
#1번 문제
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

imp=10000

sampling_count=[50,100,200]

plt.figure(figsize=(15,5))

for i in range(3):
    
    cur=np.array([])
    for _ in range(imp):
        now=0
        for j in range(sampling_count[i]):
            if np.sum(np.random.choice(range(1,46),6,replace=False))==80:
                now+=1
        
        if now>100:
            now=100
        cur=np.append(cur,now)
    plt.subplot(1,3,i+1)
    hist_bin=np.arange(0,101,1) #10 이상은 거의 나오지 않는다.
    # hist_bin=np.arange(0,11,1) #이걸로 하면 편하게 볼수있다.
    plt.hist(cur,bins=hist_bin)
    plt.title("sampling_count="+str(sampling_count[i]))
    plt.xlabel("count")
    plt.ylabel("frequency")
plt.tight_layout()
plt.show()
```

```python
#2번 문제
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


imp=10000

sampling_count=[50,100,200]

plt.figure(figsize=(15,5))

for i in range(3):
    
    cur=np.array([])
    for _ in range(imp):
        now=0
        for j in range(sampling_count[i]):
            a=np.random.choice(45,6,replace=False)
            for k in a:
                if 0<k<7:
                    now+=1
                    break # 하나라도 1~6으로 나온다면 체크를 중단한다
        
        cur=np.append(cur,now)
    plt.subplot(1,3,i+1)
    plt.hist(cur,bins=20)
    plt.title("sampling_count="+str(sampling_count[i]))
    plt.xlabel("count")
    plt.ylabel("frequency")
plt.tight_layout()
plt.show()
```
